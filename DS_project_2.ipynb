```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Titanic dataset.csv")
```

```python
data.isnull().sum() #Finding NAN values
```

```python
data.dropna(subset=["Embarked"], inplace=True) #Removal of NAN Values
data["Cabin"].fillna("Unknown", inplace=True)
data["Age"].fillna(data["Age"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(6, 5))
sns.histplot(data["Age"], kde=True)
plt.title("Demographic Dividend")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=data, x="Sex", hue="Age")
plt.title("Active Entities(Distribution by Sex)")
plt.xlabel("Sex")
plt.ylabel("Individual")
plt.show()
```

```python
plt.hist(data['Fare'], bins=20)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Fare Distribution')
plt.show()
```
