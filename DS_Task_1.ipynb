```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/world_population.csv')
df.head()
```

```python
df_final = df.drop(['Country/Territory',	'Capital',	'Continent'], axis=1)
```

```python
chosen_country = 'Afghanistan'

# Filter the original DataFrame 'df' before dropping columns
country_data = df[df['Country/Territory'] == chosen_country]
country_data = country_data.drop(['Country/Territory',	'Capital',	'Continent', 'Area (km²)', 'Density (per km²)','World Population Percentage'], axis=1)

# Extract years from column names, removing ' Population' and converting to integers, excluding 'Growth Rate'
years = [int(col.replace(' Population', '')) for col in country_data.columns[1:] if 'Population' in col]
# Adjust slicing to match the number of year columns
population = country_data.iloc[:, 2:10].values.flatten()

plt.figure(figsize=(3, 5))
plt.bar(years, population, color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')

plt.show()
```
